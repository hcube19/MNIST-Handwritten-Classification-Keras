```python
import numpy as np
import pandas as pd
from keras.datasets import mnist
```

```python
#load Data
(xtrain,ytrain),(xtest,ytest)=mnist.load_data()
```

```python
print(xtrain.shape,"     ",xtest.shape)
```

```python
#convert pixels from int to float
#Normalize the images

xtrain=xtrain.astype(float)/255
xtest=xtest.astype(float)/255
```

```python
#resize the dimensions

xtrain.resize((60000,28,28,1))
xtest.resize((10000,28,28,1))
```

```python
print(xtrain.shape,"     ",xtest.shape)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
model=Sequential()
```

```python
model.add( Conv2D( 32,(3,3), input_shape=(28,28,1), activation="relu" ) )
model.add( MaxPooling2D( (2,2) ) )
```

```python
model.add( Conv2D( 64,(3,3), activation="relu" ) )
model.add( MaxPooling2D( (2,2) ) )
```

```python
model.add( Conv2D( 16,(3,3), activation="relu" ) )
model.add( MaxPooling2D( (2,2) ) )
```

```python
model.add( Flatten() )
```

```python
model.add( Dense(128,activation="relu") )
```

```python
model.add(Dense(10,activation="softmax"))
```

```python
model.compile( metrics=["accuracy"], optimizer="adam", loss="categorical_crossentropy")
```

```python
model.summary()
```

```python
#convert labels from 0-1 into a sparse type matrix label encode

ytrain = to_categorical(ytrain, 10)
ytest = to_categorical(ytest, 10)
```

```python
model.fit(xtrain, ytrain, batch_size=20, epochs=20, validation_split=0.3, steps_per_epoch=20)
```

```python
from keras.models import save_model

save_model(model,"MNIST.h5")
```

```python
test=xtest[:9950]
test_label=ytest[:9950]
```

```python
model.evaluate(test,test_label)
```

```python
y=model.predict(test)
```

```python
correct=0
wrong=0
for i in range(len(test)):
  label=np.where(test_label[i]==1)
  prediction=np.where(y[i]==np.max(y[i]))
  print("LABEL: ",label," PREDICTED: ",prediction)

  if label==prediction:
    correct=correct+1
  else:
    wrong=wrong+1

print("CORRECT: ",correct," WRONG: ",wrong)
```
